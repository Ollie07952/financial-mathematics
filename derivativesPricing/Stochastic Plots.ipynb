```python
import matplotlib.pyplot as plt
from derivativesPricing.stochasticProcesses import gbm

walks = gbm(177.82, 0.05, 0.36, 25, 76)
fig, ax = plt.subplots()
ax.plot(walks.T, linewidth = 1)
ax.set(xlabel = "Day",ylabel = "Stock Price", title = "Geometric Brownian Motion")
ax.xaxis.tick_bottom();
```
